```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("train_csv2.csv")
train.head()
```

```python
val = pd.read_csv("validation_csv2.csv")
val.head()
```

```python
test = pd.read_csv("test_csv2.csv")
test.head()
```

```python
X_train = train.drop(['PassengerId','Survived'],axis=1)
```

```python
y_train = train['Survived']
```

```python
X_val = val.drop(['PassengerId','Survived'],axis=1)
```

```python
y_val = val['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression(solver= "lbfgs",max_iter=1000)
```

```python
logModel.fit(X_train,y_train)
```

```python
logModel.score(X_train,y_train)
```

```python
logModel.coef_
```

```python
logModel.intercept_
```

# Evaluation of val

```python
X_val.head()
```

```python
y_val_predict = logModel.predict(X_val)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_val,y_val_predict))
```

```python
logModel.score(X_train,y_train)
```

# Prediction of test dataset

```python
X_test = test.drop(["PassengerId"],axis=1)
```

```python
y_test_predict = logModel.predict(X_test)
```

```python
AllSub = pd.DataFrame({ 'PassengerId': test['PassengerId'],
                        'Survived': y_test_predict
    
})

AllSub.to_csv("Second_submission_titanic_logistic.csv",index= False)
```
